```python
magnetometer_frame.columns = ["time", "x", "y", "z"]
gyroscope_frame.columns = ["time", "wx", "wy", "wz"]
```

```python
from numpy import arctan2, pi

@np.vectorize
def caculate_theta(x, y):
    return arctan2(y, x);

@np.vectorize
def caculate_gamma(x, y, z):
    return arctan2(z, np.sqrt(x * x + y * y));
```
